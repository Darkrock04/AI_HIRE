# Basic Generative Models: Autoencoders

While models like CNNs are *Discriminative* (they predict a label for an image), **Generative Models** attempt to understand the underlying distribution of the data so they can *create entirely new data* (e.g., generating a completely new face).

## Autoencoders
An Autoencoder is a neural network designed to learn a compressed representation of data.
- **The Encoder:** Takes high-dimensional input (like a 784-pixel image) and compresses it through a bottleneck layer into a tiny "latent vector" (e.g., 32 numbers).
- **The Decoder:** Takes that tiny 32-number vector and tries to reconstruct the original 784-pixel image.
- **The Goal:** Train the network to make the output image identical to the input image. If it succeeds, it means the 32-number vector perfectly captures the "essence" of the image.

*(Note: Basic autoencoders aren't great at generating *new* data, which is why Variational Autoencoders (VAEs) and GANs were invented!)*

```python
import torch.nn as nn

class SimpleAutoencoder(nn.Module):
    """Compresses data and then rebuilds it."""
    def __init__(self, input_dim=784, hidden_dim=32):
        super(SimpleAutoencoder, self).__init__()
        
        # Compress to bottleneck
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, hidden_dim)
        )
        
        # Decompress back to original size
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim),
            nn.Sigmoid() # Scale outputs between 0 and 1
        )
        
    def forward(self, x):
        latent = self.encoder(x)
        reconstructed = self.decoder(latent)
        return reconstructed
```

# Generative Models: Autoencoders


```python
import torch.nn as nn
```

```python
# 1. Autoencoder Architecture
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        # Encoder compresses a 784 pixel image down to 3 numbers (latent space)
        self.encoder = nn.Sequential(
            nn.Linear(784, 128),
            nn.ReLU(),
            nn.Linear(128, 3)
        )
        # Decoder tries to recreate the 784 pixel image from those 3 numbers
        self.decoder = nn.Sequential(
            nn.Linear(3, 128),
            nn.ReLU(),
            nn.Linear(128, 784),
            nn.Sigmoid()
        )

    def forward(self, x):
        latent = self.encoder(x)
        reconstructed = self.decoder(latent)
        return reconstructed
```
