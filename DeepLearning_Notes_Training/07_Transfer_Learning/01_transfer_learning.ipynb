# Transfer Learning & Fine-Tuning

Training massive CNNs or Transformers from scratch requires millions of images/texts, hundreds of GPUs, and weeks of time. **Transfer Learning** is the industry standard shortcut.

## The Theory of Transfer Learning

Instead of starting with randomized weights, you download a model (like ResNet) that has already been trained on a massive dataset (like ImageNet, which has 14 million images).

1. **Freeze the Base:** The early layers of the model have already learned how to detect edges, shapes, and textures. You "freeze" these weights so they don't change.
2. **Replace the Head:** You remove the final classification layer (which predicts 1000 ImageNet categories) and replace it with a new layer for your specific task (e.g., predicting 2 categories: Cat or Dog).
3. **Train the Head:** You train only the new final layer on your small dataset. The model uses its massive pre-learned knowledge to instantly master your specific task!

```python
import torch
import torch.nn as nn
import torchvision.models as models

def setup_transfer_learning(num_custom_classes=2):
    """Downloads a pre-trained ResNet18 and modifies it for a custom task."""
    
    # 1. Download pre-trained model
    model = models.resnet18(pretrained=True)
    
    # 2. Freeze all the base layers (so we don't destroy their pre-learned knowledge)
    for param in model.parameters():
        param.requires_grad = False
        
    # 3. Replace the final layer (the "head")
    # ResNet18's final layer is called 'fc' and takes 512 input features
    num_features = model.fc.in_features
    model.fc = nn.Linear(num_features, num_custom_classes)
    
    # Now, only model.fc has requires_grad=True. 
    # When you train, it will only update this final layer!
    return model
```
