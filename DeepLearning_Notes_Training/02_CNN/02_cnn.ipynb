# Convolutional Neural Networks (CNN)

**CNNs** are the standard neural network architecture for **Computer Vision** (image recognition, object detection).

## Why not use an ANN for images?
If you flatten a 1000x1000 color image into a 1D array, you get 3,000,000 input neurons. A standard ANN would need billions of weights for a single layer, leading to massive overfitting and computational impossibility. Also, flattening destroys the 2D spatial relationships of pixels (e.g., an eye is a circle of pixels next to each other).

## Core Concepts of a CNN:
1. **Convolutional Layers:** Instead of connecting every input neuron to every output neuron, a CNN slides a small "filter" (e.g., 3x3 pixels) over the image. This filter acts as a feature detector (looking for edges, curves, or textures).
2. **Pooling Layers (e.g., MaxPooling):** Shrinks the image representation to reduce computational load and prevent overfitting, while keeping the most important features.
3. **Fully Connected (Dense) Layers:** After extracting all the features using Conv and Pool layers, the final output is flattened and fed into a standard ANN to make the final classification (e.g., "Cat" vs "Dog").

```python
import torch.nn as nn

class BasicCNN(nn.Module):
    """A standard CNN architecture for image classification."""
    def __init__(self, num_classes=10):
        super(BasicCNN, self).__init__()
        # Convolutional Layer: Detects features
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        # Pooling Layer: Downsamples the image
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        # Fully Connected Layer: Makes the final decision
        self.fc = nn.Linear(16 * 16 * 16, num_classes) # Assuming 32x32 input image
        
    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool(x)
        # Flatten the 2D matrices into a 1D array for the Linear layer
        x = x.view(x.size(0), -1) 
        x = self.fc(x)
        return x
```

# Computer Vision: CNN


```python
import torch.nn as nn
```

```python
# 1. Define a CNN
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Conv2d extracts features (edges, textures)
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        # MaxPool reduces image size to save compute
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(16 * 14 * 14, 10) # Assuming 28x28 input image

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = x.view(-1, 16 * 14 * 14) # Flatten
        x = self.fc1(x)
        return x

print('CNN architecture defined!')
```
