# Advanced Generative Models: GANs & Diffusion

Basic Autoencoders struggle to generate *sharp*, realistic new data. To solve this, researchers invented GANs and Diffusion Models, which currently power state-of-the-art systems like Midjourney and DALL-E.

## Generative Adversarial Networks (GANs)

**Theory:**
A GAN pits two neural networks against each other in a zero-sum game:
1. **The Generator:** Its job is to create fake images from random noise.
2. **The Discriminator:** Its job is to look at images and guess if they are real (from the training dataset) or fake (from the Generator).

- **The Dynamic:** As they train together, the Discriminator gets better at spotting fakes, which forces the Generator to get better at creating incredibly realistic images to fool it.

```python
import torch
import torch.nn as nn

class GAN(nn.Module):
    """A conceptual wrapper for a GAN architecture."""
    def __init__(self):
        super(GAN, self).__init__()
        
        # Generator: Takes random noise and outputs an image (e.g., 28x28 = 784)
        self.generator = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(),
            nn.Linear(256, 784),
            nn.Tanh() # Scales pixel values
        )
        
        # Discriminator: Takes an image (784) and outputs a probability of being Real (1) or Fake (0)
        self.discriminator = nn.Sequential(
            nn.Linear(784, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )
```

## Diffusion Models

**Theory:**
Diffusion Models take a completely different mathematical approach and are generally more stable to train than GANs.
1. **Forward Process (Destruction):** You take a real image and slowly add Gaussian noise to it over hundreds of steps until it's just pure, unrecognizable static.
2. **Reverse Process (Creation):** You train a neural network (usually a U-Net) to *predict and remove* the noise. 
- Once trained, you can give it pure static, and it will step-by-step "denoise" it into a brand new, realistic image.
