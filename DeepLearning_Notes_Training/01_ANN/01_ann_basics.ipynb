# Deep Learning Basics: ANN
Building a simple feed-forward neural network using PyTorch.


```
import torch
import torch.nn as nn
import torch.optim as optim
```

```
# 1. Define the Neural Network Architecture
class SimpleANN(nn.Module):
    def __init__(self):
        super(SimpleANN, self).__init__()
        self.fc1 = nn.Linear(10, 50) # Input layer (10 features) to Hidden Layer (50 neurons)
        self.relu = nn.ReLU()        # Activation function to introduce non-linearity
        self.fc2 = nn.Linear(50, 1)  # Hidden Layer to Output Layer (1 neuron)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

model = SimpleANN()
print(model)
```
