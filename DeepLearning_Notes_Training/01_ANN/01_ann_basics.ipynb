# Artificial Neural Networks (ANN)

**Deep Learning** is a subset of Machine Learning based on Artificial Neural Networks (ANNs), inspired by the structure of the human brain. While classic ML struggles with raw, unstructured data (like images or text), Deep Learning thrives on it.

## Core Concepts of an ANN:
1. **Neurons & Layers:** An ANN consists of an Input Layer, one or more Hidden Layers, and an Output Layer. Each layer contains "neurons" that hold a mathematical value.
2. **Weights & Biases:** The connections between neurons have weights (which multiply the input) and biases (which shift the output). These are the actual variables the network "learns".
3. **Activation Functions:** Without activation functions (like ReLU or Sigmoid), a neural network is just a giant linear regression model. Activation functions inject *non-linearity*, allowing the network to learn complex patterns.
4. **Forward Propagation:** Passing data from the input layer through the network to get a prediction.
5. **Backpropagation:** The "learning" phase. The network calculates how wrong its prediction was (the Loss), and uses Calculus (Chain Rule) to send that error backwards through the network, updating the weights to make the prediction better next time.

```python
import torch
import torch.nn as nn

class SimpleANN(nn.Module):
    """A basic Artificial Neural Network with one hidden layer in PyTorch."""
    def __init__(self, input_size, hidden_size, num_classes):
        super(SimpleANN, self).__init__()
        # Linear transformation: y = xA^T + b
        self.layer1 = nn.Linear(input_size, hidden_size)  
        # Activation function: max(0, x)
        self.relu = nn.ReLU()                             
        # Output layer
        self.layer2 = nn.Linear(hidden_size, num_classes) 
        
    def forward(self, x):
        # Forward propagation path
        out = self.layer1(x)
        out = self.relu(out)
        out = self.layer2(out)
        return out
```
