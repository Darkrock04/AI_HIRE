## Step 1: Import Required Libraries
We need tools to pad our text sequences and build LSTM networks.

```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

print("Libraries imported successfully!")
```

## Step 2: Load the IMDB Movie Review Dataset
Keras has a built-in dataset of movie reviews, already tokenized (words converted to integers).

```python
vocab_size = 10000 # Only keep the top 10,000 most frequent words
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)

print(f"Training sequences: {len(X_train)}")
print(f"Example sequence (as numbers): {X_train[0][:10]}")
```

## Step 3: Pad Sequences
Neural networks expect fixed-size inputs, but movie reviews have different lengths. We must pad them with zeros or truncate them to a fixed length.

```python
max_length = 200 # Limit reviews to 200 words
X_train_padded = pad_sequences(X_train, maxlen=max_length)
X_test_padded = pad_sequences(X_test, maxlen=max_length)

print(f"Padded training data shape: {X_train_padded.shape}")
```

## Step 4: Build the LSTM Model Architecture
Instead of flat data, we use an `Embedding` layer to turn word integers into dense vectors, then pass them to an `LSTM` which acts as our recurrent memory.

```python
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=128, input_length=max_length),
    LSTM(64, dropout=0.2, recurrent_dropout=0.2), # The core memory layer
    Dense(1, activation='sigmoid') # Binary output for Sentiment (Positive/Negative)
])

model.summary()
```

## Step 5: Compile and Train the Model
Training an LSTM takes significantly more time than a simple ANN because the data is processed sequentially over time steps.

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Note: In a real scenario, you'd use more epochs. We use 3 here for quick execution.
history = model.fit(X_train_padded, y_train, epochs=3, batch_size=64, validation_split=0.2, verbose=1)
print("LSTM Training Complete!")
```

## Step 6: Evaluate on Test Data
Finally, we see how well our memory-equipped neural network performs on unseen movie reviews.

```python
test_loss, test_acc = model.evaluate(X_test_padded, y_test, verbose=0)
print(f"Test Accuracy: {test_acc*100:.2f}%")
```
