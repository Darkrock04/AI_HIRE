# Sequence Data: RNNs and LSTMs


```python
import torch.nn as nn
```

```python
# 1. Define an LSTM
class TimeSeriesLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(TimeSeriesLSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # LSTM outputs the full sequence and the hidden states
        out, (hn, cn) = self.lstm(x)
        # We only care about the final prediction step usually
        out = self.fc(out[:, -1, :]) 
        return out
```

# Recurrent Neural Networks (RNN) & LSTMs

Standard ANNs and CNNs have a major flaw: they have **no memory**. They treat every input as independent. This fails for **Sequential Data** like text (sentences) or Time Series (stock prices), where the order of data matters.

## Recurrent Neural Networks (RNNs)
RNNs solve this by looping the output of a neuron back into itself. As it processes a sentence word by word, it maintains a "hidden state" (a memory) of the words it saw previously.
- **The Problem:** The Vanishing Gradient Problem. During backpropagation on long sentences, the math causes the memory of early words to fade to zero. RNNs have severe "short-term memory."

## Long Short-Term Memory (LSTM)
LSTMs are a specialized type of RNN designed to fix the short-term memory problem. They introduce a "cell state" (a conveyor belt of long-term memory) and use mathematical **Gates** (Forget Gate, Input Gate, Output Gate) to explicitly learn what information to remember and what to forget.

```python
import torch.nn as nn

class BasicLSTM(nn.Module):
    """An LSTM model for sequence processing (e.g., text sentiment analysis)."""
    def __init__(self, input_size, hidden_size, num_classes):
        super(BasicLSTM, self).__init__()
        # The LSTM layer automatically handles the complex gates and memory
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        # The final decision layer
        self.fc = nn.Linear(hidden_size, num_classes)
        
    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_size)
        # The LSTM returns the output sequence and the final hidden/cell states
        out, (hidden_state, cell_state) = self.lstm(x)
        
        # We typically only care about the very last output of the sequence for classification
        final_output = out[:, -1, :]
        return self.fc(final_output)
```
