# Chat Summarization Middleware
LLMs have token limits (e.g., 8k or 128k). If a conversation goes on for hours, passing the entire history will cause an error (and cost a fortune). 

Summarization middleware intercepts the memory, automatically summarizes older messages, and only passes the summary to the LLM.


```
from langchain.memory import ConversationSummaryMemory
from langchain_openai import ChatOpenAI
from langchain.chains import ConversationChain
```

```
# 1. Setup Summarization Middleware
# We use an LLM specifically to summarize the history in the background.
llm = ChatOpenAI(temperature=0)
summary_memory = ConversationSummaryMemory(llm=llm)

# 2. Attach to a Chain
# conversation = ConversationChain(
#     llm=llm,
#     memory=summary_memory,
#     verbose=True # We turn this on so you can physically see the middleware rewriting the memory!
# )
```
