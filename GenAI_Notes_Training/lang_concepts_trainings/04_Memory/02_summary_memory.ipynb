```python
from langchain.memory import ConversationSummaryMemory
from langchain_groq import ChatGroq

def setup_summary_memory():
    # Use a small, cheap model just for summarizing the background chat
    summary_llm = ChatGroq(model='llama3-8b-8192')
    memory = ConversationSummaryMemory(llm=summary_llm)
    return memory
```
