## 01 LCEL InMemory History

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import InMemoryChatMessageHistory

# 1. Setup the LLM
llm = ChatGroq(model='llama-3.3-70b-versatile')

# 2. Setup the Memory Object
memory = InMemoryChatMessageHistory()

# 3. Create a Prompt that specifically includes a MessagesPlaceholder for the history
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{question}")
])

# 4. Define the execution function
def chat_with_memory(query: str):
    # Format the prompt by passing the current history and the new question
    formatted_prompt = prompt.invoke({
        "chat_history": memory.messages,
        "question": query
    })
    
    # Add the formatted Human message to memory
    memory.add_messages(formatted_prompt.messages)
    
    # Get the AI response
    response = llm.invoke(formatted_prompt)
    
    # Add the AI response to memory
    memory.add_message(response)
    
    return response.content

# Test it out!
print("Turn 1:", chat_with_memory("Hi, my name is Alice!"))
print("Turn 2:", chat_with_memory("What is my name?"))
```
