## 01 Conversation Buffer

```python
from langchain.memory import ConversationBufferMemory

def setup_memory():
    memory = ConversationBufferMemory(return_messages=True)
    memory.chat_memory.add_user_message("Hi, my name is Alice!")
    memory.chat_memory.add_ai_message("Hello Alice, how can I help you?")
    return memory.load_memory_variables({})
```

## 02 Summary Memory

```python
from langchain.memory import ConversationSummaryMemory
from langchain_groq import ChatGroq

def setup_summary_memory():
    # Use a small, cheap model just for summarizing the background chat
    summary_llm = ChatGroq(model='llama3-8b-8192')
    memory = ConversationSummaryMemory(llm=summary_llm)
    return memory
```
