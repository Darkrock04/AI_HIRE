## 🔑 Step 2: Load API Key from .txt File

```python
# Load the Groq API key from a .txt file
with open("groq_api_key.txt", "r") as f:
    api_key = f.read().strip()

print(" Groq API Key loaded!")
print(f" Starts with: {api_key[:8]}...")
```

## 🤖 Step 3: Initialize the Model using LangChain + Groq


```python
from langchain_groq import ChatGroq

# Initialize LLaMA 3 via Groq
llm = ChatGroq(
    model="llama-3.3-70b-versatile",   # Free open-source LLaMA 3 model
    groq_api_key=api_key,
    temperature=0.7
)

print(" LLaMA 3.3 70B ready via LangChain + Groq!")
```

## ▶️ Step 4: Invoke the Model and Get Output

```python
# Simple invoke — send a prompt, get a response
response = llm.invoke("What is LangChain? Explain in 3 simple lines.")

print(" Response from LLaMA 3:")
print("-" * 40)
print(response.content)
```

## 📝 Step 5: Try More Prompts

```python
response2 = llm.invoke("Write a one-liner joke about Python programming.")
print(response2.content)
```

```python
response3 = llm.invoke("Explain what an LLM is to a 10-year-old.")
print(response3.content)
```

## 🧾 Step 6: Inspect the Full Response Object

```python
response_full = llm.invoke("What is 2 + 2? Answer in one word.")

print(" Full AIMessage object:")
print(response_full)

print("\n Just the text:")
print(response_full.content)

print("\n Metadata (tokens used, model, etc.):")
print(response_full.response_metadata)
```

##  Bonus: Other Free Models Available on Groq


```python
#  Try switching models — just change the model name below!

# Available free models on Groq:
# "llama-3.3-70b-versatile"     ← Best quality (default)
# "llama3-8b-8192"              ← Smaller, faster
# "mixtral-8x7b-32768"          ← Mixtral open-source
# "gemma2-9b-it"                ← Google Gemma 2

llm2 = ChatGroq(
    model="gemma2-9b-it",   # Switching to Gemma 2
    groq_api_key=api_key,
    temperature=0.5
)

response_gemma = llm2.invoke("Tell me a fun fact about space.")
print(" Gemma 2 says:")
print(response_gemma.content)
```
