## Imports

```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import InMemoryChatMessageHistory
```

## Initialize LLM and Memory

```python
os.environ['GROQ_API_KEY'] = 'gsk_YOUR_API_KEY_HERE'
llm = ChatGroq(model='llama-3.3-70b-versatile')
memory = InMemoryChatMessageHistory()
```

## Create Prompt Template

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert Data Science Mentor. Answer clearly."),
    MessagesPlaceholder(variable_name="memory"),
    ("human", "{question}")
])
```

## Chatbot Function

```python
def chatbot(query):
    msgs = prompt.invoke({"memory": memory.messages, "question": query})
    memory.add_messages(msgs.messages)
    response = llm.invoke(msgs)
    memory.add_message(response)
    return response.content
```

## Test Chatbot

```python
answer1 = chatbot("What is a Neural Network?")
answer2 = chatbot("Can you give me an example of that?")
```
