## **What is LangChain?**
**LangChain** is an open-source framework designed to simplify the creation of applications using Large Language Models (LLMs).

By itself, an LLM (like GPT-4, LLaMA, or Claude) is just a text-completion engine. It takes text in, and spits text out. It doesn't have memory, it can't search the internet, and it doesn't know about your private company PDFs.

In short:
> **LangChain is the "glue" that connects LLMs to the outside world.**

It provides standard, reusable components so you don't have to write thousands of lines of code just to let an AI read a PDF or remember what you said 5 minutes ago.

## **Key Components of LangChain**

### **1. Models (I/O)**
The core engine. LangChain standardizes the way you talk to ANY model. Whether you are using a local model running on your laptop or a cloud model, the code looks exactly the same.

### **2. Prompts (Templates)**
Instead of hardcoding strings, you create templates with variables. Example: `"Translate the following {language} sentence to English: {sentence}"`

### **3. Chains & LCEL**
You rarely just send a raw prompt to an LLM. Usually, you format a prompt, pass it to the model, and then parse the output. LangChain Expression Language (LCEL) lets you chain these like this: 
`chain = prompt | model | output_parser`

### **4. Memory**
LLMs are stateless (they have goldfish memory). LangChain provides `ConversationBufferMemory` to automatically inject the history of your chat into every new prompt behind the scenes.

### **5. Retrieval (RAG)**
The most popular use case. LangChain provides Document Loaders (to read PDFs), Text Splitters (to cut them into chunks), and Integrations with Vector Databases (like ChromaDB) to easily build "Chat with your Data" applications.
