## 01 Llms And Chat Models

```python
import os
from langchain_groq import ChatGroq

def query_model(query: str):
    os.environ['GROQ_API_KEY'] = 'gsk_YOUR_API_KEY'
    llm = ChatGroq(model='llama-3.3-70b-versatile', temperature=0)
    response = llm.invoke(query)
    return response.content
```

## 02 Prompt Templates

```python
from langchain_core.prompts import ChatPromptTemplate

def create_chat_prompt(topic: str):
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a helpful assistant. Keep answers under 2 sentences."),
        ("human", "Tell me about {topic}.")
    ])
    return prompt.format_messages(topic=topic)
```

## 03 Output Parsers

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field

class Joke(BaseModel):
    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")

def parse_joke_output(llm_raw_response: str):
    parser = JsonOutputParser(pydantic_object=Joke)
    return parser.invoke(llm_raw_response)
```
