## 01 Llms And Chat Models

```python
import os
from langchain_groq import ChatGroq

os.environ['GROQ_API_KEY'] = 'gsk_YOUR_API_KEY'
llm = ChatGroq(model='llama-3.3-70b-versatile', temperature=0)

response = llm.invoke("What is the capital of France?")
print(response.content)
```

## 02 Prompt Templates

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Keep answers under 2 sentences."),
    ("human", "Tell me about {topic}.")
])

messages = prompt.format_messages(topic="Machine Learning")
response = llm.invoke(messages)
print(response.content)
```

## 03 Output Parsers

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field

class Joke(BaseModel):
    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")

parser = JsonOutputParser(pydantic_object=Joke)

# We use LCEL (pipe operator) to connect the prompt, llm, and parser
joke_prompt = ChatPromptTemplate.from_template("Tell me a joke about {topic}.\n{format_instructions}")

chain = joke_prompt | llm | parser

result = chain.invoke({
    "topic": "programming", 
    "format_instructions": parser.get_format_instructions()
})

print(type(result)) # It is a parsed python dictionary!
print(result)
```
