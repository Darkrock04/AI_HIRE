## 01 Langfuse Tracing

```python
import os

def setup_tracing():
    # When using LangSmith or Langfuse, you typically just set environment variables.
    # LangChain automatically detects them and logs all traces.
    os.environ["LANGCHAIN_TRACING_V2"] = "true"
    os.environ["LANGCHAIN_API_KEY"] = "ls_YOUR_KEY"
```

## 02 Streaming Callbacks

```python
def stream_response(chain, query):
    # Using .stream() instead of .invoke() yields chunks in real-time
    for chunk in chain.stream({"input": query}):
        print(chunk, end="", flush=True)
```
