# Hybrid Retrieval (BM25 + Semantic Search)


```
from langchain.retrievers import EnsembleRetriever
from langchain_community.retrievers import BM25Retriever
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
```

```
# 1. Setup Data
texts = ["Apples are red.", "Apples are green.", "Bananas are yellow."]
```

```
# 2. BM25 Keyword Retriever
# Excellent for exact word matches (e.g., finding the exact ID 'AX-409')
bm25_retriever = BM25Retriever.from_texts(texts)
bm25_retriever.k = 2
```

```
# 3. Vector Retriever (Semantic Search)
# Excellent for understanding meaning (e.g., matching 'fruit' to 'Apples')
vectorstore = Chroma.from_texts(texts, OpenAIEmbeddings())
vector_retriever = vectorstore.as_retriever(search_kwargs={"k": 2})
```

```
# 4. Ensemble Retriever (Hybrid)
# Uses Reciprocal Rank Fusion to combine the best of both worlds.
ensemble_retriever = EnsembleRetriever(
    retrievers=[bm25_retriever, vector_retriever], weights=[0.5, 0.5]
)
# docs = ensemble_retriever.get_relevant_documents("yellow fruit")
# print(docs)
```
