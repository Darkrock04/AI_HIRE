# Fine-Tuning: LoRA and PEFT
Concepts on how to train a massive 70B parameter model on a single GPU using Low-Rank Adaptation.


```
# 1. Why LoRA?
# Standard fine-tuning updates every single weight in the neural network (too expensive).
# LoRA freezes the main weights and only trains a tiny 'adapter' matrix on the side.
# This reduces memory usage by 90%.
```
