# Prompt Engineering Techniques


```
from langchain.prompts import PromptTemplate, FewShotPromptTemplate
```

```
# 1. Few-Shot Prompting
# Give the LLM examples so it learns the pattern before answering.
examples = [
    {"word": "happy", "antonym": "sad"},
    {"word": "tall", "antonym": "short"}
]
example_template = PromptTemplate(input_variables=["word", "antonym"], template="Word: {word}\nAntonym: {antonym}")

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_template,
    prefix="Give the antonym of every input\n",
    suffix="Word: {input}\nAntonym:",
    input_variables=["input"]
)
# print(few_shot_prompt.format(input="big"))
```

```
# 2. Chain of Thought (CoT)
# Forcing the LLM to 'think out loud' drastically improves math and logic accuracy.
cot_prompt = """Question: I had 5 apples. I ate 2. I bought 10 more. How many do I have?
Let's think step by step.
"""
```
