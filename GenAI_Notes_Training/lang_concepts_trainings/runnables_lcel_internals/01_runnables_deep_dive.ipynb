# Deep Dive into the Runnable Protocol
Master the inner workings of LCEL.


```
from langchain_core.runnables import RunnablePassthrough, RunnableParallel, RunnableLambda
```

```
# 1. RunnableLambda
# Converts any standard Python function into an LCEL-compatible Runnable.
def add_five(x: int) -> int:
    return x + 5

runnable_func = RunnableLambda(add_five)
# print(runnable_func.invoke(10)) # Returns 15
```

```
# 2. RunnableParallel
# Runs multiple runnables at the exact same time.
parallel_chain = RunnableParallel({
    "math_branch": runnable_func,
    "string_branch": RunnableLambda(lambda x: f"The number is {x}")
})
# print(parallel_chain.invoke(10))
```
