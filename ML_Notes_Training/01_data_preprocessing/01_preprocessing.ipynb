# Exhaustive Data Preprocessing
Learn how to clean raw tabular data using pandas and Scikit-Learn.


```
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```
# 1. Handling Missing Data
# We use SimpleImputer to fill missing numeric values with the mean, and categorical values with the most frequent.
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')
```

```
# 2. Feature Scaling & Encoding
# StandardScaler ensures all numerical features have a mean of 0 and std of 1.
# OneHotEncoder converts categories into binary columns.
scaler = StandardScaler()
encoder = OneHotEncoder(handle_unknown='ignore')
```

```
# 3. Building a Preprocessing Pipeline
# Pipelines ensure we apply the exact same transformations to our training and testing data without data leakage.
preprocessor = ColumnTransformer(transformers=[
    ('num', Pipeline(steps=[('imputer', num_imputer), ('scaler', scaler)]), ['age', 'income']),
    ('cat', Pipeline(steps=[('imputer', cat_imputer), ('encoder', encoder)]), ['city', 'profession'])
])
print('Preprocessor ready!')
```
