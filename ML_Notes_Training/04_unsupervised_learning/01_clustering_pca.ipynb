# Unsupervised Learning: Clustering & PCA

**Unsupervised Learning** deals with *unlabeled data*. The model is given a dataset without answers (no `y`), and its job is to discover hidden patterns, structures, or groupings within the input data (`X`).

## 1. K-Means Clustering

**Theory:**
K-Means tries to group data into `K` distinct clusters based on similarity (distance).
1. Randomly place `K` "centroids" (cluster centers) in the data space.
2. Assign every data point to the closest centroid.
3. Recalculate the centroids to be the exact middle of the newly formed clusters.
4. Repeat until the centroids stop moving.

```python
from sklearn.cluster import KMeans

def perform_kmeans(X, n_clusters=3):
    """Groups unlabeled data into K clusters and returns the cluster labels."""
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    return kmeans.cluster_centers_, cluster_labels
```

## 2. Principal Component Analysis (PCA)

**Theory:**
PCA is a **Dimensionality Reduction** technique. If your dataset has 100 features (dimensions), it's hard to visualize and slow to train on.
- PCA mathematically compresses the dataset down to a smaller number of "Principal Components" (e.g., 2 or 3 components) while preserving as much of the original data's *variance* (information) as possible.
- **Use Cases:** Data visualization (compressing to 2D), speeding up training, removing noise.

```python
from sklearn.decomposition import PCA

def compress_data_pca(X, n_components=2):
    """Compresses a high-dimensional dataset down to n_components."""
    pca = PCA(n_components=n_components)
    compressed_X = pca.fit_transform(X)
    
    # Shows how much information was preserved (e.g. 95%)
    variance_preserved = sum(pca.explained_variance_ratio_)
    return compressed_X, variance_preserved
```
