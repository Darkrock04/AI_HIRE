## Step 1: Import Required Libraries
We need Pandas for our data, Matplotlib for graphs, and Scikit-Learn for our ML algorithm.

```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

print("Libraries imported successfully!")
```

## Step 2: Load the Kaggle Dataset
Change the `file_path` variable below to perfectly match the exact name of the file you uploaded.

```python
# IMPORTANT: Change 'Salary_Data.csv' to the exact name of your downloaded Kaggle file!
file_path = 'Salary_Data.csv'

# Load the data
df = pd.read_csv(file_path)

print(f"Dataset Size: {df.shape[0]} Rows, {df.shape[1]} Columns")
display(df.head())
```

## Step 3: Extract the X (Inputs) and y (Answers)
We tell the computer what column it is studying (`X`), and what column it is trying to predict (`y`).

*Note: If your Kaggle dataset has different column names, change `YearsExperience` and `Salary` below to match your dataset exactly!*

```python
# Change these column names if your dataset uses different words!
X = df[['YearsExperience']].values 
y = df['Salary'].values

print("X and y variables successfully extracted!")
```

## Step 4: The Golden Rule - Train / Test Split
We lock 20% of the data away in a test vault. The ML model is never allowed to see this until the very end.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,       # 20% goes into the vault
    random_state=42      # Ensures consistent results on multiple runs
)

print(f"Training Data Size: {len(X_train)} samples")
print(f"Test Data Size (In Vault): {len(X_test)} samples")
```

## Step 5: Initialize and Train the Machine Learning Model
The simple `fit(X, y)` command triggers the AI to analyze the data and calculate the mathematical rules.

```python
# 1. Create the empty AI brain
model = LinearRegression()

# 2. Teach the AI the rules!
model.fit(X_train, y_train)

print("Model Training Complete!")
print(f"Base Starting Salary found by AI: ${model.intercept_:.2f}")
print(f"Value of one year of experience found by AI: ${model.coef_[0]:.2f}")
```

## Step 6: Test the AI on the Unseen Vault Data
We unleash the AI on the 20% Test data and compare its guesses side-by-side with reality.

```python
# Ask the AI to predict unseen salaries
y_pred = model.predict(X_test)

# Compare predictions vs reality
results_df = pd.DataFrame({
    'Years Experience': X_test.flatten(),
    'Actual Reality Salary ($)': y_test,
    'AI Predicted Salary ($)': y_pred.round(2)
})

display(results_df)

# Show the final accuracy percentage
accuracy = model.score(X_test, y_test)
print(f"\nAI Accuracy Score: {accuracy * 100:.1f}%")
```

## Step 7: Visualize the AI's Brain
Let's put the original data (Blue Dots) on a graph, and see the literal line that the AI decided was the absolute smartest pattern (Red Line).

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Real Employees')
plt.plot(X, model.predict(X), color='red', linewidth=3, label='AI Extracted Pattern')
plt.title('AI Brain Analysis vs Real Data', fontsize=16)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Final Step: Use the AI to predict the future!
A brand new candidate asks for a job. They have **6.5 years** of experience. What should you pay them based on the total company data?

```python
# The brackets [[ ]] mean it is asking the 2D matrix model to predict a new value
new_amount_of_experience = [[6.5]]

predicted_offer = model.predict(new_amount_of_experience)

print(f"For {new_amount_of_experience[0][0]} years of experience...")
print(f"The AI mathematically suggests an offering salary of: ${predicted_offer[0]:,.2f}")
```

## 🛑 DEPLOYMENT STEP: Export the Brain!
If we close this tab, the AI dies forever. To deploy this to the real world (using FastAPI & Streamlit), we must freeze the model's brain into a file called a `.pkl` (Pickle) object.

```python
# Save the model to a file
joblib.dump(model, 'salary_model.pkl')
print("Model perfectly frozen and saved to 'salary_model.pkl'!")
print("WARNING: Open the Colab folder sidebar, download 'salary_model.pkl' to your computer, and drop it into your local backend folder!")
```
