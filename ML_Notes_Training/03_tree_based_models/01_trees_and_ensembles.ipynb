# Tree-Based Models & Ensembles
Implementation of Decision Trees, Random Forests, and Gradient Boosting.


```
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```
# 1. Decision Tree (Prone to overfitting)
tree = DecisionTreeClassifier(max_depth=5)
```

```
# 2. Random Forest (Bagging)
# Creates 100 decision trees and averages their predictions to prevent overfitting.
rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```
# 3. Gradient Boosting (Boosting)
# Creates trees sequentially, where each new tree corrects the errors of the previous ones.
gb = GradientBoostingClassifier(learning_rate=0.1, n_estimators=100)
```
