# Tree-Based Models & Ensembles

Decision Trees are intuitive models that split data like a flowchart. **Ensemble methods** combine multiple weak models (like many trees) to create one highly accurate, robust "super model".

## 1. Decision Trees

**Theory:**
A Decision Tree splits data based on questions (e.g., "Is Age > 30?"). 
- **Gini Impurity / Entropy:** The mathematical metrics used to decide which feature to split on. The tree always picks the feature that creates the most "pure" child nodes (nodes where most samples belong to a single class).
- **Pros/Cons:** Highly interpretable, but extremely prone to **overfitting** (memorizing the training data instead of generalizing).

```python
from sklearn.tree import DecisionTreeClassifier

def train_decision_tree(X_train, y_train):
    """Trains a basic Decision Tree."""
    model = DecisionTreeClassifier(max_depth=5) # Limit depth to prevent overfitting
    model.fit(X_train, y_train)
    return model
```

## 2. Random Forests (Bagging Ensemble)

**Theory:**
A Random Forest solves the Decision Tree's overfitting problem using **Bagging** (Bootstrap Aggregating).
1. It builds hundreds of independent Decision Trees.
2. Each tree is trained on a random subset of the data and a random subset of the features.
3. **Voting:** For a final prediction, all trees "vote" (majority wins for classification, average for regression). This drastically reduces variance.

```python
from sklearn.ensemble import RandomForestClassifier

def train_random_forest(X_train, y_train):
    """Trains a Random Forest with 100 trees."""
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    return model
```

## 3. Gradient Boosting (Boosting Ensemble)

**Theory:**
Unlike Random Forests (which build trees independently in parallel), **Boosting** builds trees *sequentially*.
1. Tree 1 makes predictions and inevitably makes some errors.
2. Tree 2 is built specifically to correct the errors of Tree 1.
3. Tree 3 corrects the errors of Tree 2, and so on.
- **Result:** A highly accurate model, but prone to overfitting if not tuned carefully.

```python
from sklearn.ensemble import GradientBoostingClassifier

def train_gradient_boosting(X_train, y_train):
    """Trains a Gradient Boosting model sequentially."""
    model = GradientBoostingClassifier(learning_rate=0.1, n_estimators=100)
    model.fit(X_train, y_train)
    return model
```
