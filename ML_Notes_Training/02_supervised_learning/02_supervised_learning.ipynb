# Supervised Learning: Regression & Classification

**Supervised Learning** is the branch of Machine Learning where we train a model using *labeled data*. This means every training example has an input (X) and a known correct output (y). The model learns the mapping from X to y so it can predict y for new, unseen data.

There are two main types of Supervised Learning:
1. **Regression:** Predicting a *continuous number* (e.g., House Price, Temperature, Stock Value).
2. **Classification:** Predicting a *category or class* (e.g., Spam/Not Spam, Cat/Dog, Benign/Malignant).

## 1. Linear Regression

**Theory:**
Linear Regression attempts to model the relationship between two variables by fitting a linear equation (a straight line) to observed data. 
- The formula is `y = mx + b` (where `m` is the slope/weight, and `b` is the intercept/bias).
- **Goal:** Minimize the Mean Squared Error (MSE), which is the average squared distance between the predicted points on the line and the actual data points.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def train_linear_regression(X_train, y_train, X_test, y_test):
    """Trains a Linear Regression model and returns the MSE."""
    model = LinearRegression()
    model.fit(X_train, y_train)          # Train the model
    
    predictions = model.predict(X_test)  # Predict continuous numbers
    error = mean_squared_error(y_test, predictions)
    return model, error
```

## 2. Logistic Regression (Classification)

**Theory:**
Despite its name, Logistic Regression is used for *classification*, not regression. 
- It uses a **Sigmoid function** to squeeze the output of a linear equation into a probability score between 0 and 1.
- If the probability > 0.5, it predicts Class 1 (e.g., Spam). If < 0.5, it predicts Class 0 (Not Spam).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

def train_logistic_regression(X_train, y_train, X_test, y_test):
    """Trains a Logistic Regression classifier and returns accuracy."""
    model = LogisticRegression()
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)  # Predict classes (0 or 1)
    accuracy = accuracy_score(y_test, predictions)
    return model, accuracy
```

## 3. Support Vector Machines (SVM)

**Theory:**
SVMs are powerful classifiers that find the "optimal hyperplane" (a boundary) that separates different classes with the maximum margin.
- **Support Vectors:** The data points closest to the hyperplane. They are the most critical points because they define the boundary.
- **Kernel Trick:** If the data isn't linearly separable, SVMs use mathematical "kernels" (like RBF) to project the data into higher dimensions where a straight boundary can separate them.

```python
from sklearn.svm import SVC

def train_svm(X_train, y_train, X_test, y_test):
    """Trains an SVM classifier using the RBF kernel."""
    model = SVC(kernel='rbf')
    model.fit(X_train, y_train)
    
    return model, accuracy_score(y_test, model.predict(X_test))
```

# Classic ML Algorithms: k-NN & Naive Bayes

Before diving into complex neural networks, it's crucial to understand the foundational algorithms of Machine Learning. They are fast, interpretable, and often serve as the perfect baseline for any project.

## 1. k-Nearest Neighbors (k-NN)

**Theory:**
k-NN is an extremely simple algorithm that stores all available cases and classifies new cases based on a similarity measure (e.g., distance functions like Euclidean).
- **How it works:** To predict the class of a new data point, k-NN looks at its `k` closest neighbors in the training data. If `k=3`, and 2 neighbors are "Cat" while 1 is "Dog", it predicts "Cat" via majority vote.
- **Pros/Cons:** No actual "training" phase (fast to train), but making a prediction is slow because it has to calculate the distance to *every* training point.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def train_knn(X_train, y_train, X_test, y_test, k=3):
    """Predicts using the k-Nearest Neighbors algorithm."""
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    return model, accuracy_score(y_test, predictions)
```

## 2. Naive Bayes

**Theory:**
Naive Bayes is a classification technique based on Bayes' Theorem with an assumption of independence among predictors. In simple terms, a Naive Bayes classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature.
- **How it works:** It calculates the probability of each class given the input features and outputs the class with the highest probability.
- **Use Cases:** It is incredibly fast and particularly effective for text classification (like Spam Filtering or Sentiment Analysis).

```python
from sklearn.naive_bayes import GaussianNB

def train_naive_bayes(X_train, y_train, X_test, y_test):
    """Predicts using the Gaussian Naive Bayes algorithm."""
    model = GaussianNB()
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    return model, accuracy_score(y_test, predictions)
```

# NLP Foundations: Bag of Words & TF-IDF

Machine Learning algorithms only understand numbers, not text. Therefore, Natural Language Processing (NLP) fundamentally involves converting raw text into numerical vectors that models can process.

## 1. Bag of Words (CountVectorizer)

**Theory:**
The Bag of Words model creates a vocabulary of all unique words in your dataset. It then represents each sentence by counting how many times each word from the vocabulary appears in it.
- **Drawback:** It treats all words equally. Frequent words like "the" or "and" dominate the counts, overshadowing the rare words that actually carry the core meaning of the sentence.

```python
from sklearn.feature_extraction.text import CountVectorizer

def apply_bag_of_words(corpus):
    """Converts a list of text sentences into word frequency counts."""
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(corpus)
    
    return vectorizer.get_feature_names_out(), X.toarray()
```

## 2. TF-IDF (Term Frequency-Inverse Document Frequency)

**Theory:**
TF-IDF solves the Bag of Words problem by assigning a weight to each word. 
- **Term Frequency (TF):** How often a word appears in a specific document.
- **Inverse Document Frequency (IDF):** Reduces the weight of words that appear in *every* document (like "the") and increases the weight of rare, specific words.
- **Result:** A matrix where highly meaningful words have high mathematical values.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def apply_tfidf(corpus):
    """Converts text into TF-IDF weighted numerical vectors."""
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(corpus)
    
    return vectorizer.get_feature_names_out(), X.toarray()
```

# Recommendation Systems: Collaborative & Content-Based Filtering

Recommendation systems are the algorithms powering Netflix, Amazon, and YouTube. They analyze user behavior and item features to suggest what a user might like next.

## 1. Content-Based Filtering

**Theory:**
Content-based filtering recommends items that are similar to those a user has liked in the past. 
- If you watch a Sci-Fi movie starring Keanu Reeves, it will recommend other Sci-Fi movies or movies with Keanu Reeves.
- **Mechanism:** It relies on item metadata (genres, actors, descriptions) and calculates similarity (e.g., Cosine Similarity) between items.

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_content_based(item_features_matrix, target_item_index):
    """Returns the similarity scores of all items compared to the target item."""
    # Calculate cosine similarity between the target item and all other items
    similarities = cosine_similarity(item_features_matrix[target_item_index:target_item_index+1], item_features_matrix)
    return similarities[0] # Returns a list of similarity scores (1.0 is a perfect match)
```

## 2. Collaborative Filtering

**Theory:**
Collaborative Filtering ignores item metadata entirely. Instead, it looks at user behavior.
- If User A and User B both highly rated Movies 1, 2, and 3, the algorithm assumes they have similar tastes.
- Therefore, if User A rates Movie 4 highly, the algorithm will recommend Movie 4 to User B, even if it knows nothing about the movie's genre or actors.

```python
import numpy as np

def user_based_collaborative_filtering(user_item_matrix, target_user_index):
    """Finds users with similar rating patterns to the target user."""
    # Calculates cosine similarity between the target user's ratings and all other users' ratings
    user_similarities = cosine_similarity(user_item_matrix[target_user_index:target_user_index+1], user_item_matrix)
    
    # Get the index of the most similar user (excluding themselves)
    user_similarities[0][target_user_index] = -1 # Prevent matching with self
    most_similar_user = np.argmax(user_similarities[0])
    
    return most_similar_user
```
